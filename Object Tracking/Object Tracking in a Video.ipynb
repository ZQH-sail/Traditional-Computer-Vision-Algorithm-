```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
tracker = cv2.TrackerMIL_create()
tracker_type = 'MIL'
```

```python
video = cv2.VideoCapture('car_moving.mp4')
print(type(video))
if not video.isOpened():
    print("Could not open Video File ")
```

```python
ok, frame = video.read()
if not ok: 
    print("Cannot read video file \n")
```

```python
# Exploring Frames
type(frame)
np.shape(frame)
#print(ok)
```

```python
bbox = cv2.selectROI(frame)
print(bbox)
```

```python
while True:
    ok, frame = video.read()
    #print(ok)
    #print(frame)
    if not ok:
        break
    
    # Start the timer
    timer = cv2.getTickCount()
    
    # Update the tracker
    ok, bbox = tracker.update(frame)
    print(ok)
    print(bbox)
    
    if not ok:
        bbox = cv2.selectROI(frame)
        
    
    # Calculating the Frames per Second
    fps = cv2.getTickFrequency()/cv2.getTickCount() - timer
    
    # Drawing a bounding box:
    if ok:
        # Tracking success
        print(ok)
        p1 = (int(bbox[0]),int(bbox[1]))
        p2 = (int(bbox[0]+bbox[2]),int(bbox[1]+bbox[3]))
        cv2.rectangle(frame, p1, p2, (255,0,0))
    
    else:
        cv2.putText(frame, "Tracking failure detected", (500,500), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (0,0,255),2)
        
        
    cv2.putText(frame, tracker_type + "Tracker", (100,80),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.75,(0,0,255),2)
    
    
    cv2.imshow("Tracking", frame)
    
    k = cv2.waitKey(100) & 0xff
    if k == 27:
        break

rerun = input("Enter '1' for re running and '0' for exiting \n")
if (rerun==0):
    video.release()
    cv2.destroyAllWindows()
```
